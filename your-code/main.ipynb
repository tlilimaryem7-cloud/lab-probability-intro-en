## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
#Taking a white ball out.
white_balls = 10
red_balls = 12
black_balls = 8

total_balls = white_balls + red_balls + black_balls

prob_white = white_balls / total_balls

print("Probability of drawing a white ball:", prob_white)
```

```python
#Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one
prob_white_first = white_balls / total_balls
total_balls -= 1  # one ball removed
white_balls -= 1   # white ball removed
prob_black_second = black_balls / total_balls

prob_both = prob_white_first * prob_black_second

print("Probability of white first and black second (without replacement):", prob_both)
```

```python
#Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
prob_white_first = white_balls / total_balls

prob_black_second = black_balls / total_balls

prob_both = prob_white_first * prob_black_second

print("Probability of white first and black second (without replacement):", prob_both)
```

```python
#Taking 3 balls and them being all different colours.
prob_white_first = white_balls / total_balls
prob_black_second = black_balls / total_balls
prob_red_third = red_balls / total_balls

# factoriel 3 =6  is the number of options ( red first , black second , white third or black first ,white seconde .....)

prob_with_replacement = 6 * prob_white_first * prob_black_second * prob_red_third

print("Probability of all different colors (with replacement):", prob_with_replacement)

```
